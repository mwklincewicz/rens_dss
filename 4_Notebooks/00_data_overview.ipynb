# Socio-Economic Status (SES) scores



```python
import pandas as pd
import glob

files = glob.glob("../1_Data_raw/SES_WOA scores/*.csv")

dfs = []
#1. appending data-sets of 2014 until 2023
for f in files:
    df = pd.read_csv(f, sep=";")
    dfs.append(df)

ses = pd.concat(dfs, ignore_index=True)

```

```python
ses.columns
```

```python
#checking missing values

print("Columns:", ses.columns.tolist())
print("\nDatatypes:\n", ses.dtypes)
print("\nMissing values:\n", ses.isna().sum())
print("\nGM/WK/BU structure:\n", ses["WijkenEnBuurten"].str[:2].value_counts())
print("\nSES columns:\n", [c for c in ses.columns if "GemiddeldeScore" in c])
print("\nrows per year:\n", ses.groupby(ses["Perioden"].str[:4]).size())
print("\nmunicipalities per year:\n", ses.groupby(ses["Perioden"].str[:4])["WijkenEnBuurten"].nunique())
```

```python
# filter municipalities
ses_gm = ses[ses["WijkenEnBuurten"].str.startswith("GM")]

# extrapolate Year (jaar) out of the column 'Perioden'
ses_gm["Jaar"]= ses_gm["Perioden"].str[:4].astype(int)

# Select SES-scores ("GemiddeldeScore_29)
ses_gm["SES_score"] = ses_gm["GemiddeldeScore_29"]

# Clean municipality codes
ses_gm["GemeenteCode"]= ses_gm["WijkenEnBuurten"].str[2:].astype(int)


ses_gm.head()
```

```python
# Only keep the useful columns
ses_gm_clean = ses_gm[["GemeenteCode","Jaar", "SES_score"]].copy()
ses_gm_clean.head()
```

```python
# Check for municipality code of "Haaren" which was split in 2021, if there is no output, the dataset has taken splitting and merging of municipalities taken into account
ses_gm_clean[ses_gm["GemeenteCode"] == 788].groupby("Jaar").size()

ses_gm_clean.groupby("Jaar")["GemeenteCode"].nunique()
```

```python
#sorting the ses_scores in the same format as the 'leegstandsmonitor'
ses_sorted = ses_gm_clean.sort_values(["Jaar", "GemeenteCode"], ascending=[False, True])
ses_sorted.head()
```

```python
# SES scores to data_intermediate
ses_sorted.to_csv("../2_Data_intermediate/ses_woa_clean.csv", index=False)

pd.read_csv("../2_Data_intermediate/ses_woa_clean.csv")
```
